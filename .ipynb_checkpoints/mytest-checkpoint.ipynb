```python
def abl2rgb(a,b,L=50):
    dimim = 1
    
    out = np.ones([dimim,dimim,3])
    abl = skimage.color.lab2rgb([L,a,b])
    print(abl)
    out[:,:,0] = np.ones([dimim,dimim])*abl[0]
    out[:,:,1] = np.ones([dimim,dimim])*abl[1]
    out[:,:,2] = np.ones([dimim,dimim])*abl[2]
    
    print(out)
    #plt.imshow(out)
```

```python
def rgb(r,g,b):
    out = np.ones([8,8,3])
    out[:,:,0] = out[:,:,0]*r
    out[:,:,1] = out[:,:,1]*g
    out[:,:,2] = out[:,:,2]*b
    
    #plt.imshow(out)
    return out
    
def findlab(r,g,b):
    im = rgb(r,g,b)
    imlab = skimage.color.rgb2lab(im)
    L = skimage.color.rgb2lab(im)[0][0][0]
    a = skimage.color.rgb2lab(im)[0][0][1]
    b = skimage.color.rgb2lab(im)[0][0][2]
    
    return (L,a,b)
```

```python
def lab(L,a,b):
    out = np.ones([8,8,3])
    out[:,:,0] = out[:,:,0]*L
    out[:,:,1] = out[:,:,1]*a
    out[:,:,2] = out[:,:,2]*b
    
    return(out)
    
def findrgb(L,a,b):
    im = lab(L,a,b)
    imrgb = skimage.color.lab2rgb(im)
    r = imrgb[0][0][0]
    g = imrgb[0][0][1]
    b = imrgb[0][0][2]

    return (r,g,b)
```

```python
print(findlab(1,0,0))
print(findrgb(53.240587943744885, 80.09230822569225, 67.20275104442868))
```

```python
L = 50

def maplab2rgb(a,b,L=L):
    return(np.asarray(findrgb(L,a,b)))

H = 220//5
W = 220//5
rgbmap = np.zeros([H,W,3])

for h in range(H):
    a = -110 + h*5
    for w in range(W):
        b = -110 + w*5
        rgbmap[h,w] = maplab2rgb(a,b)
        check = findlab(rgbmap[h,w][0],rgbmap[h,w][1],rgbmap[h,w][2])[0]
        tol = .005
        if(check < L - tol or check > L + tol):
            if L < 40:
                rgbmap[h,w] = [1,1,1]
            else:
                rgbmap[h,w] = [0,0,0]
                
weights = np.zeros([H,W,3])

                
plt.imshow(rgbmap)
```
